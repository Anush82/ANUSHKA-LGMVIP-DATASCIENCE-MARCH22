## IMAGE TO PENCIL SKETCH WITH PYTHON
## LEVEL- BEGINNER

### IMPORTING NECESSARY LIBRARIES 

```python
import cv2
import matplotlib.pyplot as plt
```

### IMPORTING ORIGINAL IMAGE 

```python
original=cv2.imread(r"C:\Users\Prachi\Pictures\bird.png")
color=cv2.cvtColor(original,cv2.cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.axis("off")
plt.imshow(color)
plt.title("ORIGINAL IMAGE")
plt.show()
```

### CONVERTING IMAGE INTO GREY SCALE IMAGE 

```python
grey_image=cv2.cvtColor(original,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.axis('off')
plt.title("GREYSCALE IMAGE")
plt.imshow(grey_image,cmap='gray')
plt.show()
```

### INVERTING THE GREY SCALE IMAGE 

```python
inverted_image=255-grey_image
plt.figure(figsize=(10,10))
plt.axis('off')
plt.title('INVERTED IMAGE')
plt.imshow(inverted_image,cmap='gray')
plt.show()
```

### BLURRING THE IMAGE USING GAUSSIAN FUNCTION 

```python
blurred=cv2.GaussianBlur(inverted_image,(21,21),0)
plt.figure(figsize=(10,10))
plt.axis('off')
plt.title('BLURRED IMAGE')
plt.imshow(blurred,cmap='gray')
plt.show()
```

### INVERTING BLURRED IMAGE(PENCIL SKETCH) 

```python
invert_blurred=255-blurred
pencil=cv2.divide(grey_image,invert_blurred,scale=256.0)
plt.figure(figsize=(10,10))
plt.axis('off')
plt.title("PENCIL IMAGE")
plt.imshow(pencil,cmap='gray')
plt.show()
```

### COMPARING ORIGINAL IMAGE WITH PENCIL SKETCH 

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(color)
plt.axis('off')
plt.title("ORIGINAL IMAGE")
plt.subplot(1,2,2)
plt.imshow(pencil,cmap='gray')
plt.axis('off')
plt.title("PENCIL SKETCH")
```
